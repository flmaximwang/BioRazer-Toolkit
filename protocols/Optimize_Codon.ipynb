```python
from dnachisel import *

protein_sequence = "MGDSMKLVMLLLKKAVTLAKLNNDDMVALEIERAAKQIVLALAVNKSDEMAKVMLALAKAVLLAAKNNDDEVAREIARAAAEIVEALKENNSDEMAKVMLALAKAVLLAAKNNDDEVAREIARAAAEIVEALRENNSDEMAKKMLELAKRVLDAAKNNDDETAREIARQAAEEVEADREN"
dna_sequence = reverse_translate(protein_sequence=protein_sequence, table="Bacterial")
problem = DnaOptimizationProblem(
    sequence=dna_sequence,
    constraints=[
        EnforceTranslation(),
        EnforceGCContent(mini=0.3, maxi=0.6, window=30),
        UniquifyAllKmers(k=9),
        AvoidPattern("XhoI_site")
    ],
    objectives=[
        CodonOptimize(species="e_coli", method="match_codon_usage"),
    ]
)
problem.resolve_constraints()
problem.optimize()
print("Original sequence:", dna_sequence)
print("Optimized sequence:", problem.sequence)
```

```python
print(problem.all_constraints_pass())
```
