```python
from dnachisel import *

original_sequence = "ATGAAAGGTGATACTAAATTAATTAATGCTTTAAATGCTTTATTAGGTTTAGCTTTAGTTGCTATTAATACTTTATTTTTAGCTGCTCGTATGGCTAAAAATTTAGGTTTAAAATTAGTTAATGATTTATTATATCATGCTTCTATTATTGCTATGAAAATTGCTGATATTTTAATTGAATTAATTTTATTTTTAGAAGGTTTACCTAATTTACAAGATTTAGGTAAATTAAATATTTCTTCTGGTTCTGGTGGTTTAATTAATGCTATGAATGCTGCTTTAGGTGTTTTATTAGTTATGATTAATCAAGCTTTTTTATTAGCTCGTATGGCTAAAAATAAAGGTGATAAATTAACTAATGATTTATTATATCATGCTTCTATTTTAGCTATGAAAGCTGCTGATTTATTAATTGAAGCTATTTTATTTGCTGAAGGTTTACCTAATTTACAAGATTTAGGTAAATTAAATATT"
problem = DnaOptimizationProblem(
    sequence=original_sequence,
    constraints=[
        EnforceTranslation(),
        EnforceGCContent(mini=0.3, maxi=0.6, window=30),
        UniquifyAllKmers(k=7)
    ]
)
problem.resolve_constraints()
problem.optimize()
print("Original sequence:", original_sequence)
print("Optimized sequence:", problem.sequence)
```
